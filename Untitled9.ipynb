```python
import shutil
import xml.etree.ElementTree as ET

def organize_dataset(image_dir="Images", target_dir="sorted_data"):
    os.makedirs(target_dir, exist_ok=True)

    for file in os.listdir(image_dir):
        if file.endswith(".xml"):
            # Parse XML to get class name
            xml_path = os.path.join(image_dir, file)
            tree = ET.parse(xml_path)
            class_name = tree.find('object').find('name').text

            # Create class folder if not exists
            class_dir = os.path.join(target_dir, class_name)
            os.makedirs(class_dir, exist_ok=True)

            # Move both image and XML
            base_name = os.path.splitext(file)[0]
            img_file = f"{base_name}.jpg"

            shutil.move(
                os.path.join(image_dir, img_file),
                os.path.join(class_dir, img_file)
            )
            shutil.move(
                os.path.join(image_dir, file),
                os.path.join(class_dir, file)
            )

organize_dataset()
print("Organization complete!")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create data generators with augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Using 20% for validation
)

# Flow from directory
train_generator = train_datagen.flow_from_directory(
    'sorted_data',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    'sorted_data',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model

# Load pre-trained model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze base layers
base_model.trainable = False

# Add custom head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(len(class_names), activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_data=val_generator,
    validation_steps=val_generator.samples // val_generator.batch_size,
    epochs=20,
    verbose=1
)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from imblearn.keras import balanced_batch_generator
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Get all image paths and labels
filepaths = []
labels = []
for class_name in os.listdir("sorted_data"):
    class_dir = os.path.join("sorted_data", class_name)
    for f in os.listdir(class_dir):
        if f.endswith(".jpg"):
            filepaths.append(os.path.join(class_dir, f))
            labels.append(class_name)

# Convert to DataFrame for easier handling
df = pd.DataFrame({'filename': filepaths, 'class': labels})

# Convert labels to numerical indices
class_names = sorted(df['class'].unique())
label_to_index = {name: i for i, name in enumerate(class_names)}
df['class_idx'] = df['class'].map(label_to_index)

# Stratified train-test split
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df['class'],
    random_state=42
)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import os

# 1. Prepare and load the dataset
filepaths = []
labels = []
for class_name in os.listdir("sorted_data"):
    class_dir = os.path.join("sorted_data", class_name)
    for f in os.listdir(class_dir):a
        if f.endswith(".jpg"):
            filepaths.append(os.path.join(class_dir, f))
            labels.append(class_name)

# Create DataFrame
df = pd.DataFrame({'filename': filepaths, 'class': labels})

# 2. Train-Validation Split
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df['class'],
    random_state=42
)

# 3. Prepare Data Generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

# 4. Calculate class weights
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(train_df['class']),
    y=train_df['class']
)
class_weights = dict(enumerate(class_weights))

# 5. Create generators
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col='filename',
    y_col='class',
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    shuffle=True
)

val_generator = val_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col='filename',
    y_col='class',
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    shuffle=False
)

# 6. Build and compile model
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(len(train_generator.class_indices), activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 7. Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_data=val_generator,
    validation_steps=val_generator.samples // val_generator.batch_size,
    epochs=50,
    class_weight=class_weights,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=3)
    ]
)
```

```python
# Evaluate the model on validation data
val_loss, val_accuracy = model.evaluate(val_generator)
print(f"Validation Accuracy: {val_accuracy*100:.2f}%")
print(f"Validation Loss: {val_loss:.4f}")
```

```python
import os
from PIL import Image
import pandas as pd

def get_dataset_stats(folder_path="sorted_data"):
    stats = []
    for class_name in os.listdir(folder_path):
        class_dir = os.path.join(folder_path, class_name)
        if os.path.isdir(class_dir):
            images = [f for f in os.listdir(class_dir) if f.endswith(('.jpg', '.png'))]
            sample_img = os.path.join(class_dir, images[0]) if images else None

            # Get image dimensions
            width, height = 0, 0
            if sample_img:
                with Image.open(sample_img) as img:
                    width, height = img.size

            stats.append({
                "Class": class_name,
                "Image Count": len(images),
                "Resolution": f"{width}x{height}",
                "Sample Image": sample_img
            })

    return pd.DataFrame(stats)

dataset_report = get_dataset_stats()
dataset_report.to_csv("dataset_report.csv", index=False)
dataset_report
```

```python
import matplotlib.pyplot as plt

def visualize_samples(folder_path="sorted_data", samples_per_class=2):
    plt.figure(figsize=(15, 20))
    for i, class_name in enumerate(os.listdir(folder_path)):
        class_dir = os.path.join(folder_path, class_name)
        images = [f for f in os.listdir(class_dir) if f.endswith(('.jpg', '.png'))][:samples_per_class]

        for j, img_name in enumerate(images):
            img_path = os.path.join(class_dir, img_name)
            plt.subplot(len(os.listdir(folder_path)), samples_per_class, i*samples_per_class + j + 1)
            img = plt.imread(img_path)
            plt.imshow(img)
            plt.title(f"{class_name}\n{img.shape[1]}x{img.shape[0]}")
            plt.axis('off')

    plt.tight_layout()
    plt.savefig("dataset_samples.jpg")
    plt.show()

visualize_samples()
```
